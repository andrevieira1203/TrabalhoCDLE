```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("tabelafinal5.csv")
df
```

```python
diagnosis_gender_count = df.groupby(['DIAGNOSES', 'GENDER']).size().unstack(fill_value=0)

# Configurando o gráfico
plt.figure(figsize=(10, 6))
diagnosis_gender_count.plot(kind='bar', stacked=True)

# Adicionando títulos e rótulos
plt.title('Relação entre DIAGNOSES e GENDER')
plt.xlabel('DIAGNOSES')
plt.ylabel('Contagem de GENDER')
plt.xticks(rotation=0)
plt.legend(title='GENDER')

# Mostrando o gráfico
plt.show()
```

```python
# Contando a quantidade de cada gênero para cada diagnóstico
diagnosis_gender_count = df.groupby(['DIAGNOSES', 'GENDER']).size().unstack(fill_value=0)

# Configurando o gráfico
plt.figure(figsize=(10, 6))
diagnosis_gender_count.plot(kind='bar', stacked=False)

# Adicionando títulos e rótulos
plt.title('Relação entre DIAGNOSES e GENDER')
plt.xlabel('DIAGNOSES')
plt.ylabel('Contagem de GENDER')
plt.xticks(rotation=0)
plt.legend(title='GENDER')

# Mostrando o gráfico
plt.show()
```

```python
diagnosis_gender_count = df.groupby(['DIAGNOSES', 'GENDER']).size().unstack(fill_value=0)

# Calculando as porcentagens
diagnosis_gender_percent = diagnosis_gender_count.div(diagnosis_gender_count.sum(axis=1), axis=0) * 100

# Configurando o gráfico
plt.figure(figsize=(10, 6))
diagnosis_gender_percent.plot(kind='bar', stacked=False)

# Adicionando títulos e rótulos
plt.title('Porcentagem de GENDER para cada DIAGNOSES')
plt.xlabel('DIAGNOSES')
plt.ylabel('Porcentagem de GENDER')
plt.xticks(rotation=0)
plt.legend(title='GENDER')

# Mostrando o gráfico
plt.show()
```

```python
# Contando a frequência de cada procedimento para cada doença
procedures = ['3893', '966', '9604', '9904', '9671', '3891', '0', '9672', '3961', '8856', '3615']
procedure_count = df.groupby('DIAGNOSES')[procedures].sum()

# Configurando o gráfico
procedure_count.plot(kind='bar', stacked=True, figsize=(12, 8))

# Adicionando títulos e rótulos
plt.title('Frequência de Procedimentos por Doença')
plt.xlabel('Doença')
plt.ylabel('Frequência dos Procedimentos')
plt.legend(title='Procedimentos')

# Mostrando o gráfico
plt.show()
```

```python
# Convertendo o DataFrame para o formato longo para facilitar o plot
df_melted = df.melt(id_vars=['DIAGNOSES'], value_vars=['3893', '966', '9604', '9904', '9671', '3891', '0', '9672', '3961', '8856', '3615'], 
                    var_name='Procedimento', value_name='Feito')

# Configurando o FacetGrid
g = sns.FacetGrid(df_melted, col='DIAGNOSES', col_wrap=2, height=5, sharex=False, sharey=False)
g.map_dataframe(sns.barplot, x='Procedimento', y='Feito', estimator=sum, ci=None)

# Adicionando títulos e rótulos
g.set_axis_labels('Procedimento', 'Número de Feitos')
g.set_titles('Doença: {col_name}')
plt.xticks(rotation=90)

# Mostrando o gráfico
plt.show()
```

```python
# Definindo os intervalos de idade
bins = [0, 9, 19, 29, 39, 49, 59, 69, 79, 89, 99]
labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99']

# Criando uma nova coluna com os grupos de idade
df['AGE_GROUP'] = pd.cut(df['AGE'], bins=bins, labels=labels, right=False)

# Calculando a média de LOS para cada grupo de idade
age_group_avg_los = df.groupby('AGE_GROUP')['LOS'].mean().reset_index()

# Configurando o gráfico
plt.figure(figsize=(12, 8))
sns.barplot(x='AGE_GROUP', y='LOS', data=age_group_avg_los)

# Adicionando títulos e rótulos
plt.title('Média do Tempo de Permanência no Hospital (LOS) por Grupo de Idade')
plt.xlabel('Grupo de Idade')
plt.ylabel('Tempo Médio de Permanência no Hospital (LOS)')
plt.xticks(rotation=45)

# Mostrando o gráfico
plt.show()
```

```python
# Definindo os intervalos de idade
bins = [0, 9, 19, 29, 39, 49, 59, 69, 79, 89, 99]
labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99']

# Criando uma nova coluna com os grupos de idade
df['AGE_GROUP'] = pd.cut(df['AGE'], bins=bins, labels=labels, right=False)

# Calculando a contagem de diagnósticos para cada grupo de idade
age_diagnosis_counts = df.groupby(['AGE_GROUP', 'DIAGNOSES']).size().reset_index(name='counts')

# Configurando o gráfico de barras empilhadas
plt.figure(figsize=(12, 8))
sns.barplot(x='AGE_GROUP', y='counts', hue='DIAGNOSES', data=age_diagnosis_counts.astype({'AGE_GROUP': 'str', 'DIAGNOSES': 'str'}))

# Adicionando títulos e rótulos
plt.title('Contagem de Diagnósticos por Grupo de Idade')
plt.xlabel('Grupo de Idade')
plt.ylabel('Contagem de Diagnósticos')
plt.xticks(rotation=45)
plt.legend(title='Diagnósticos')

# Mostrando o gráfico
plt.show()
```
